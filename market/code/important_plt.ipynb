```python
import matplotlib.pyplot as plt
data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True) #! sakhte plot setaee
axs[0].bar(names, values) #! bar plot
axs[0].grid() #! baraye gride baiad az axs estefade koni
fig.supylabel('values', fontsize=15) #! baraye label bayad az fig va sup estefade koni
axs[1].grid() #! baraye gride baiad az axs estefade koni
axs[1].scatter(names, values) #! rasme scatter plot
fig.supxlabel('name',fontsize=15) #! baraye label bayad az fig va sup estefade koni
axs[2].plot(names, values) #! rasme plot
axs[2].grid() #! grid
fig.suptitle('Categorical Plotting') #! nahve title dar in plot fargh dare
plt.savefig('save', bbox_inches= 'tight', pad_inches=1) #! save
```
