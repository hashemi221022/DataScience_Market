```python
df5 = df4.dropna() #! hazf dade nan
```

```python
df5.describe(include='all') #! describe be sorate koli
```

```python
df5['Adj Close'] = df5['Adj Close'].astype(int) #! tabdile flot to int
df5
```

```python
Year = pd.to_datetime(df5.Date).dt.year #! tabdile date be year
```

```python
#plt.figure(figsize=(5,3), dpi=800) #! keyfiat va size koli nemodar
plt.title('price of market in 1965 t0 2021', fontsize=20) #! title
plt.plot(Year, df5['Adj Close']) #! rasme plot
#plt.stem(Year, df5['Adj Close'])
plt.xlabel('DATE', fontsize=15) #! lebel x
plt.xticks(rotation=90) #! charkhesh 90 daraje
plt.ylabel('Adj Close', fontsize=15) #! lebel y
plt.grid() #! rasme grid
plt.savefig('market', bbox_inches= 'tight', pad_inches=1) #! save kardan plot
plt.show()
```
